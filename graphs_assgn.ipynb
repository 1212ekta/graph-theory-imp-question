# Write a code to find the degree of each vertex, and store it in a dictionary.<br>Sort the keys of the dictionary by the degree stored in the values

```python
def get_sorted_degree_dict(graph):
    #Step 1 calculate degree of each vertex
    degree = {}
    for node in graph:
        count = 0
        for i in graph[node]:
            count+=1
        degree[node] = count

    #Extract the keys into a list
    keys = []
    for k in degree:
        keys+=[k]

    #step 3: Sort the keys using selection sort based on degree values
    i = 0
    while i<len(keys):
        j = i+1
        while j<len(keys):
            if degree[keys[i]]>degree[keys[j]]:
                #Swap if current vertex has higher degree
                temp = keys[i]
                keys[i] = keys[j]
                keys[j] = temp
            j+=1
        i+=1
    #step 4:create a new dictionary with sorted keys
    sorted_degree_dict = {}
    for k in keys:
        sorted_degree_dict[k] = degree[k]
    return sorted_degree_dict

graph = {"S1":["S2","S3"],"S2":["S1","S4"],"S3":["S1"],"S4":["S2"]}
result = get_sorted_degree_dict(graph)
print("Sorted Degree :")
print(result)
```

<h2>Write a code to inter-convert the 3 graph representations we have learnt.


```python
# Function to convert an adjacency list to an adjacency matrix
def adjacency_matrix(data, directed=False):
    # Step 1: Get all unique node names and sort them manually
    nodes = []
    for key in data:
        nodes.append(key)

    # Manual sorting (bubble sort style) 
    for i in range(len(nodes)):
        for j in range(i + 1, len(nodes)):
            if nodes[i] > nodes[j]:
                temp = nodes[i]
                nodes[i] = nodes[j]
                nodes[j] = temp

    n = len(nodes)  # Total number of nodes

    # Step 2: Create an n x n matrix filled with 0s
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(0)
        matrix.append(row)

    # Step 3: Fill the matrix based on the edges in the adjacency list
    for i in range(n):  # Loop through each node
        node = nodes[i]
        neighbours = data[node]

        for k in range(len(neighbours)):
            neighbour = neighbours[k]

            # Find the index of the neighbour manually
            j = -1
            for idx in range(len(nodes)):
                if nodes[idx] == neighbour:
                    j = idx
                    break

            # Mark the edge from node i to j
            matrix[i][j] = 1

            # If the graph is undirected, mark the reverse edge too
            if not directed:
                matrix[j][i] = 1

    return matrix


# Function to return the same adjacency list (no change needed)
def adjacency_list(data):
    return data


# Function to convert adjacency list to edge list
def edge_list(data, directed=False):
    edges = []  # Empty list to store edge pairs

    for node in data:
        neighbours = data[node]

        for k in range(len(neighbours)):
            neighbour = neighbours[k]

            # Check if edge is already in the list (for undirected graphs)
            already_present = False
            if not directed:
                for m in range(len(edges)):
                    if edges[m][0] == neighbour and edges[m][1] == node:
                        already_present = True
                        break

            # Add edge if it's not already added (or always for directed)
            if directed or not already_present:
                edges.append((node, neighbour))

    return edges


# Sample Graph (Adjacency List Format)
connection = {
    "s1": ["s2", "s5"],
    "s2": ["s1", "s3", "s5"],
    "s3": ["s4", "s2"],
    "s4": ["s3"],
    "s5": ["s1", "s2"]
}

# Choose whether the graph is directed
is_directed = False  # Set to True if the graph is directed

# Generate all 3 representations
adj_matrix = adjacency_matrix(connection, directed=is_directed)
adj_list = adjacency_list(connection)
edge_lst = edge_list(connection, directed=is_directed)

# Print the Adjacency Matrix
print("Adjacency Matrix:")
for i in range(len(adj_matrix)):
    print(adj_matrix[i])

# Print the Adjacency List
print("\nAdjacency List:")
print(adj_list)

# Print the Edge List
print("\nEdge List:")
print(edge_lst)
```

<h2>Given a graph and two vertices, check if they are adjacent.<h2>

```python
# Function to check if two vertices are adjacent (connected by an edge)
def are_adjacent(adj_list, vertex1, vertex2, directed=False):
    # Step 1: Check if vertex1 exists in the adjacency list
    if vertex1 in adj_list:
        # Step 2: Get the list of neighbours of vertex1
        neighbours1 = adj_list[vertex1]

        # Step 3: Check if vertex2 is in the list of neighbours
        for i in range(len(neighbours1)):
            if neighbours1[i] == vertex2:
                return True  # vertex1 is connected to vertex2

    # Step 4: If the graph is undirected, also check reverse connection
    if not directed:
        # Check if vertex2 exists in the adjacency list
        if vertex2 in adj_list:
            neighbours2 = adj_list[vertex2]

            # Check if vertex1 is in the neighbours of vertex2
            for j in range(len(neighbours2)):
                if neighbours2[j] == vertex1:
                    return True  # vertex2 is connected to vertex1

    # Step 5: If no connection is found
    return False


# Example graph (adjacency list format)
connection = {
    "s1": ["s2", "s5"],
    "s2": ["s1", "s3", "s5"],
    "s3": ["s4", "s2"],
    "s4": ["s3"],
    "s5": ["s1", "s2"]
}

# Example input vertices
vertex1 = "s1"
vertex2 = "s3"

# Call the function and print result
result = are_adjacent(connection, vertex1, vertex2, directed=False)
print("Are", vertex1, "and", vertex2, "adjacent?", result)
```

<h2>Check if a graph is complete.<h2>


```python
# Function to check if the given adjacency list represents a complete graph
def complete_adj_lst(lst):
    # Step 1: Extract all nodes manually
    nodes = []
    for key in lst:
        nodes.append(key)

    n = len(nodes)  # Total number of nodes

    # Step 2: Check if each node has exactly n-1 neighbors
    for i in range(n):
        node = nodes[i]
        neighbors = lst[node]

        count = 0
        for j in range(len(neighbors)):
            count = count + 1

        if count != n - 1:
            return "Not complete"

    # Step 3: Check if every node is connected to all other nodes
    for i in range(n):
        node = nodes[i]
        neighbors = lst[node]

        for j in range(n):
            other_node = nodes[j]

            if other_node != node:
                # Check if other_node exists in neighbors
                found = False
                for k in range(len(neighbors)):
                    if neighbors[k] == other_node:
                        found = True
                        break

                if not found:
                    return "Not complete"

    # If all checks pass, the graph is complete
    return "It is complete"


# Test cases
lst1 = {
    "s1": ["s2", "s5"],
    "s2": ["s1", "s3", "s5"],
    "s3": ["s4", "s2"],
    "s4": ["s3"],
    "s5": ["s1", "s2"]
}
print(complete_adj_lst(lst1))  # Output: Not complete

lst2 = {
    "s1": ["s2", "s3"],
    "s2": ["s1", "s3"],
    "s3": ["s1", "s2"]
}
print(complete_adj_lst(lst2))  # Output: It is complete
```

<h2>Check if graph is connected<h2>

```python
def is_graph_connected(graph):
    """
    Check if a graph is connected using Depth-First Search (DFS)
    """
    
    
    
    # Handle empty graph case
    if len(graph) == 0:# If it is empty graph its already connected
        return True
    
    # Initialize visited tracking
    visited = {}
    for node in graph:
        visited[node] = False
    
    # Get first node to start DFS
    start_node = None
    for node in graph:
        start_node = node
        break
    
    # DFS implementation using recursion
    def dfs(current_node):
        visited[current_node] = True
        # Visit all neighbors
        for neighbor in graph[current_node]:
            if not visited[neighbor]:
                dfs(neighbor)
    
    # Start DFS from the first node
    dfs(start_node)
    
    # Check if all nodes were visited
    all_visited = True
    for node in visited:
        if not visited[node]:
            all_visited = False
            break
    
    return all_visited


#Example Usage
connected_graph = {
        'A': {'B': 1, 'C': 1},
        'B': {'A': 1, 'D': 1},
        'C': {'A': 1, 'D': 1},
        'D': {'B': 1, 'C': 1}
    }
    
    
disconnected_graph = {
        'A': {'B': 1},
        'B': {'A': 1},
        'C': {'D': 1},
        'D': {'C': 1}
    }
    
print("Connected Graph Test:", is_graph_connected(connected_graph))      # Should print True
print("Disconnected Graph Test:", is_graph_connected(disconnected_graph))  # Should print False
```

<h2>Given a graph and a list of vertices, check if it forms a walk, or a trail or a path or none of these.<h2>

```python
def check_sequence_type(graph, sequence):
    """
    Determine if a sequence of vertices forms a walk, trail, path, or none in a graph.
    
    """
    
    # Edge case: empty sequence or single vertex
    if len(sequence) == 0:
        return "None"
    if len(sequence) == 1:
        return "Path"  # Trivially a path (and trail and walk)
    
    # Track visited edges to check for repeated edges (for trail/path)
    visited_edges = set()
    # Track visited vertices to check for repeated vertices (for path)
    visited_vertices = set()
    
    is_walk = True
    is_trail = True
    is_path = True
    
    # Check each consecutive pair in the sequence
    for i in range(len(sequence) - 1):
        current = sequence[i]
        next_node = sequence[i + 1]
        
        # 1. Check if consecutive vertices are connected (walk requirement)
        if next_node not in graph.get(current, {}):
            return "None"
        
        # Create edge representation (undirected, sorted to avoid duplicates like A-B vs B-A)
        edge = tuple(sorted((current, next_node)))
        
        # 2. Check for repeated edges (trail requirement)
        if edge in visited_edges:
            is_trail = False
            is_path = False
        visited_edges.add(edge)
        
        # 3. Check for repeated vertices (path requirement)
        if i == 0:  # Only add first vertex initially
            visited_vertices.add(current)
        if next_node in visited_vertices:
            is_path = False
        visited_vertices.add(next_node)
    
    # Determine the most specific type
    if is_path:
        return "Path"
    elif is_trail:
        return "Trail"
    elif is_walk:
        return "Walk"
    else:
        return "None"


# Example Usage:
if __name__ == "__main__":
    # Sample graph
    graph = {
        'A': {'B', 'C'},
        'B': {'A','D'},
        'C': {'A', 'D'},
        'D': {'B', 'C', 'E'},
        'E': {'D'}
    }
    
    # Test cases
    test_sequences = [
        (['A', 'B', 'D', 'E'], "Path"),     # No repeats
        (['A', 'B', 'D', 'C', 'A'], "Trail"),  # Repeats vertex but not edges
        (['A', 'B', 'A', 'C'], "Walk"),      # Repeats edge A-B
        (['A', 'B', 'F'], "None"),           # Invalid edge
        (['A'], "Path"),                     # Single vertex
        (['A', 'B', 'D', 'C', 'A', 'B'], "Walk")  # Repeats edge and vertex
    ]
    
    for seq, expected in test_sequences:
        result = check_sequence_type(graph, seq)
        print(f"Sequence {seq}: {result} (Expected: {expected})")
```

<h2>check if a graph is tree or not<h2>

```python
def is_tree(graph):
    """
    Check if a given graph is a tree.
    
    
    """
    
    if not graph:
        return True  # Empty graph is considered a tree
    
    visited = set()
    parent = {}
    
    # Start DFS from the first node
    start_node = next(iter(graph))
    stack = [(start_node, None)]  # (current_node, parent_node)
    edge_count = 0
    
    while stack:
        current, prev = stack.pop()
        
        if current in visited:
            # If we encounter a visited node that's not our parent, cycle exists
            return False
        
        visited.add(current)
        
        for neighbor in graph.get(current, []):
            if neighbor != prev:  # Don't go back to parent
                stack.append((neighbor, current))
                edge_count += 1
    
    # A tree must have exactly (V-1) edges (since we count each edge twice)
    # and must be connected (visited all nodes)
    return len(visited) == len(graph) and edge_count == 2 * (len(graph) - 1)

    # Test cases
tree_graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D'],
        'C': ['A'],
        'D': ['B']
    }
    
cyclic_graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D'],
        'C': ['A', 'D'],
        'D': ['B', 'C']
    }
    
disconnected_graph = {
        'A': ['B'],
        'B': ['A'],
        'C': ['D'],
        'D': ['C']
    }
    
print("Tree graph:", is_tree(tree_graph))            # True
print("Cyclic graph:", is_tree(cyclic_graph))        # False
print("Disconnected graph:", is_tree(disconnected_graph))  # False
```

<h2>Given a connected cyclic graph, find its spanning tree.</h2>

```python
def find_spanning_tree(graph):
    """
    Find a spanning tree of a connected graph using DFS.
    A spanning tree is a subgraph that includes all vertices with minimal edges (no cycles).
    
    Args:
        graph (dict): Adjacency list representation of the graph
                     Format: {'A': ['B', 'C'], 'B': ['A'], 'C': ['A']}
    
    Returns:
        dict: Spanning tree as an adjacency list
    """
    
    # Handle empty graph case
    if not graph:
        return {}
    
    # Initialize spanning tree with empty neighbor lists
    spanning_tree = {}
    for node in graph:
        spanning_tree[node] = []
    
    # Track visited nodes to prevent revisiting
    visited = set()
    
    # Get the first node in the graph without using iter()
    start_node = None
    for node in graph:
        start_node = node
        break
    
    # DFS stack stores tuples of (current_node, parent_node)
    stack = [(start_node, None)]
    
    # Perform DFS to build the spanning tree
    while stack:
        # Pop the current node and its parent from stack
        current_node, parent_node = stack.pop()
        
        # Process only if not visited
        if current_node not in visited:
            # Mark current node as visited
            visited.add(current_node)
            
            # Add edge between current node and parent (if exists)
            if parent_node is not None:
                spanning_tree[parent_node].append(current_node)
                spanning_tree[current_node].append(parent_node)
            
            # Explore neighbors in reverse order for DFS
            # (This gives same result as iterating in order)
            for neighbor in reversed(graph[current_node]):
                # Skip the parent to avoid going backwards
                if neighbor != parent_node:
                    # Add to stack with current_node as parent
                    stack.append((neighbor, current_node))
    
    return spanning_tree

# Connected graph
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B'],
    'F': ['C']
}

spanning_tree = find_spanning_tree(graph)
print(f"spanning tree of the given tree is:{spanning_tree}")
```

<h2>Given a tree,find the number of leaf nodes</h2>

```python
def count_leaves(tree):
    """
    Count the number of leaf nodes in a tree.
    
    A leaf node is defined as a node with only one connection (degree 1) in an undirected tree.
    In a directed tree, it would be a node with no children (out-degree 0).
    

        
    
    Examples:
        >>> tree = {'A': ['B'], 'B': ['A', 'C'], 'C': ['B']}
        >>> count_leaves(tree)
        2  # Nodes A and C are leaves
    """
    
    leaf_count = 0  # Initialize counter
    
    # Iterate through each node in the tree
    for node in tree:
        # Get the degree of the node (number of connections)
        degree = len(tree[node])
        
        # For undirected trees, leaf nodes have degree 1
        # (connected only to their parent)
        if degree == 1:
            leaf_count += 1
    
    return leaf_count
#Example usage
tree = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}

print("The number of leaf nodes is:", count_leaves(tree))
```

<h2>Height of the tree<h2>

```python
def height_of_node(graph, node):
    """
    Returns the height of a node in a tree.
    Height of a node = max depth from this node to a leaf.
    A leaf node has height 0.
    """
    if not graph[node]:  # If the node has no children
        return 0
    else:
        
        return 1 + max(height_of_node(graph, child) for child in graph[node])
    
#Example usage   
tree = {
    'A': ['B', 'C'],
    'B': [],
    'C': ['D'],
    'D': []
}

print("Height of A:", height_of_node(tree, 'A'))  # ➤ 2
print("Height of B:", height_of_node(tree, 'B'))  # ➤ 0
print("Height of C:", height_of_node(tree, 'C'))  # ➤ 1
print("Height of D:", height_of_node(tree, 'D'))  # ➤ 0



    
```

<h2>Depth of the tree<h2>

```python
# Function to find the depth of a node in a tree
def depth_of_node(tree, current_node, target_node, depth=0):
    """
    Returns the depth of target_node in the tree.

   """

    # Base case: if the current node is the target node
    if current_node == target_node:
        return depth

    # Go through all the children (connected nodes)
    for child in tree[current_node]:
        # Make recursive call to search deeper in the tree
        d = depth_of_node(tree, child, target_node, depth + 1)

        # If the node was found, return the depth
        if d != -1:
            return d

    # If node not found in any subtree, return -1
    return -1


# Example tree (undirected tree rooted at 'A', but used as a directed tree in search)
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

# Example usage
target = 'E'
starting_node = 'A'

# Call the function
depth = depth_of_node(tree, starting_node, target)

# Print the result
if depth != -1:
    print(f"The depth of node '{target}' from root '{starting_node}' is: {depth}")
else:
    print(f"Node '{target}' not found in the tree starting from '{starting_node}'")
```
